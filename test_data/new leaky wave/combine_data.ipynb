### Combine the 3 sets into one set with default values from parametric sweep

```python
import pandas as pd
defaults = {
    "cpw_in [mm]": 2,
    "feed_l [mm]": 3.75,
    "patch_l [mm]": 3.5,
    "cpw_g [mm]": 0.18,
    "Feed_W [mm]": 1,
    "ground_w [mm]": 1,
    "patch_ground_w [mm]": 1,
    "patch_w [mm]": 4.75
}
```

```python
df1 = pd.read_csv("S11_Data_Set1.csv")
df2 = pd.read_csv("S11_Data_Set2.csv")
df3 = pd.read_csv("S11_Data_Set3.csv")
combined_df = pd.concat([df1, df2, df3])
```

```python
for k, v in defaults.items():
    combined_df[[k]] = combined_df[[k]].fillna(value=v)
```

```python
for col_name in ["Freq [GHz]", "dB(S(1,1)) []"]:
    combined_df[col_name] = combined_df.pop(col_name)
```
